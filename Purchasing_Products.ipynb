# Read Inventory

```python
import json
import codecs

records=json.load(codecs.open('record.json', 'r', 'utf-8-sig'))
```

# Purchase

```python
product_id        = input("Enter Product ID : ")
product_quantity  = input("Quantity : ")

if(int(records[product_id]['Product Quantity']) >= int(product_quantity)):
    discount=int(records[product_id]['Discount'])*int(records[product_id]['Price'])/100
    amt=(int(records[product_id]['Price'])-discount)*int(product_quantity)
    
    records[product_id]['Product Quantity'] = str(int(records[product_id]['Product Quantity']) - int(product_quantity))
              
    print('---------- BILL ----------\n')
    print('Product ID : ', product_id)
    print('Product Name : ',records[product_id]['Product Name'])
    print('Quantity : ', product_quantity+'\n')
    print('Discount : ',discount)
    dis=0
    if(amt>=1000):
              dis=0.05*discount
              print('Remark : YaY won CashBack of ',dis, ' Rupees\n')
    print('**************************')
    print('Total Amount : ',amt-dis)
    print('Money Saved : ', discount*int(product_quantity)+dis)
    print('\nThanks for Shopping!!!')
    import time

    t = time.ctime()

    print('\n------- Billing date and Time', t)
    

else:
                     print('Out of Stock')

                    
                     
js = json.dumps(records)

fd = open("record.json",'w')
fd.write(js)
fd.close()

```
